```python
#Aqui criamos um novo dataframe apenas com as colunas que eu quero..
ca_df = mercadoDeDados_df[["('P1_h ', 'Nivel de Ensino')", "('P1_a ', 'Idade')", "('P2_b ', 'Setor')", "('P2_h ', 'Faixa salarial')"]] 
display(ca_df)
```

```python
#Exibir da 10a. linha até a 20a.linha
display(ca_df.loc[9:19])
```

```python
# Criar um dataframe forma_df contendo 
# apenas as 4 colunas ()
# somente Graduação/Bacharelado exibir na tela
forma_df = ca_df.loc[ca_df["('P1_h ', 'Nivel de Ensino')"] == 'Graduação/Bacharelado']
display(forma_df)

display(forma_df["('P1_a ', 'Idade')"].max())
```

```python
# Qual a média de idade de pessoas que recebem de R$ 20.001/mês a R$ 25.000/mês no sudeste separados por pessoas que tem a formação em  graduação
display(mercadoDeDados_df.loc[(mercadoDeDados_df["('P1_e_b ', 'Regiao onde mora')"] == 'Sudeste') & 
                            (mercadoDeDados_df["('P1_h ', 'Nivel de Ensino')"] == 'Graduação/Bacharelado') & 
                            (mercadoDeDados_df["('P2_h ', 'Faixa salarial')"].isin(['de R$ 20.001/mês a R$ 25.000/mês'])), 
                            ["('P1_a ', 'Idade')"]].mean())
```

```python
# Como mostrar média de valor de idae POR SALRIO Brasil todo?
media_por_salario_df = ca_df[["('P2_h ', 'Faixa salarial')", "('P1_a ', 'Idade')"]].groupby(by="('P2_h ', 'Faixa salarial')").mean().round(2)
display(media_por_salario_df)
```

```python
#Quero adicionar uma coluna de valor booleano no mercadoDeDados_df
# chamada "Ativo" que, por padrão, vai ser True para 
# TODAS as linhas
mercadoDeDados_df['Ativo'] = True
print(mercadoDeDados_df.info())
display(mercadoDeDados_df.head())
```
